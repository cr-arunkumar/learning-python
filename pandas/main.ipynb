```python
df = pd.DataFrame({
    'name': ["John", "Alice", "Bob", "Gurif", "Jiok"],
    'age': [30, 25, 35, 40, 28],
    'city': ["New York", "San Francisco", "Los Angeles", "Tokyo", "Sydney"],
    'income': [100000, 80000, 120000, 150000, 90000]
})
```

```python
print(df)
```

```python
if 'name' in df.columns:
    df = df.set_index('name')
    print(df)
else:
    print("Error: 'name' column not found in the DataFrame.")
```

```python
print(df.loc['Alice', "age"])

# we can access the items using their index
print("using index", df.iloc[1, 1])

# we can access the items using their index
print("using index", df.loc["Alice", "age"])
```

```python
if 'name' in df.columns:
    df = df.set_index('name')
    print(df)
```

```python
print(df.loc['Alice', "age"])
print("using index", df.iloc[1, 0])
print("using at method", df.at["Alice", "age"])
print("using iat method", df.iat[1, 0])
```

```python
# we can also do slcing operation
print(df)
print("----- Slicing ------")
print("0-2", df.iloc[0:2])
print("1-3  with first column in a result set", df.iloc[1:3, 0])
```

```python

df.age = df.age*2
print(df)
```

```python
# iterating over dataframe
print("----- Iterating over dataframe ------")
for index, row in df.iterrows():
    # print(index,"---\n", row,"\n\n\n")
    pass

# iterating over columns
print("----- Iterating over columns ------")
for col in df.columns:
    # print(col,"---\n", df[col],"\n\n\n")
    pass

for i, col in df.items():
    print(i, "--->", col)

# sort the dataframe
print("----- Sorting dataframe ------")
df = df.sort_values(by=['age', 'income'], ascending=[False, True])
print(df)
```

```python
import datetime as dt
import random as rnd

df["birth_date"] = df["age"].apply(
    lambda x: dt.date.today() - dt.timedelta(days=rnd.randint(1000, 3065)))
print(df)
```

```python
# merging and joining dataframes
df1 = pd.DataFrame({
    'id': [1, 2, 3, 4, 5],
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'city': ['New York', 'San Francisco', 'Los Angeles', 'Tokyo', 'Sydney'],
    'age': [30, 25, 35, 40, 28],
    'income': [100000, 80000, 120000, 150000, 90000]
})

df2 = pd.DataFrame({
    'id': [3, 4, 5, 6, 7],
    'name': ['Charlie', 'David', 'Eve', 'Frank', 'Grace'],
    'city': ['Los Angeles', 'Tokyo', 'Sydney', 'New York', 'San Francisco'],
    'age': [35, 40, 28, 30, 25],
    'income': [120000, 150000, 90000, 100000, 80000],
    'favorite_color': ['blue', 'green', 'yellow', 'pink', 'purple']
})

# join operation
join_df = pd.concat([df1, df2], ignore_index=True)
join_df.to_html('join_result.html', index=False)
join_df
```

```python
merge_df = pd.merge(df1, df2)
merge_df
```
